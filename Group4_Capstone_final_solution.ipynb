**Time of Day and Day of Week**

```python
def getTimeCat(dateTime):
    # extract time categories
    str_time = datetime.datetime.strptime(dateTime, "%d-%m-%Y %H:%M:%S")
    ts = datetime.datetime.time(str_time)

    timecat=0
    
    # --> Slot 1 = 0400-1000
    s1Start = datetime.time(9, 15, 1)
    s1End = datetime.time(10, 15, 0)

    # --> Slot 2 = 1000-1600
    s2Start = datetime.time(10,15 , 1)
    s2End = datetime.time(11, 15, 0)

    # --> Slot 3 = 1600-2200
    s3Start = datetime.time(11, 15, 1)
    s3End = datetime.time(12, 15, 0)

    # --> Lslot 4 = 2200-0400
    s4Start = datetime.time(12, 15, 1)
    s4End = datetime.time(13, 15, 0)
    
    # --> Slot 5 = 0400-1000
    s5Start = datetime.time(13, 15, 1)
    s5End = datetime.time(14, 15, 0)

    # --> Slot 6 = 1000-1600
    s6Start = datetime.time(14, 15, 1)
    s6End = datetime.time(15, 15, 0)

    # --> Slot 7 = 1600-2200
    s7Start = datetime.time(15, 15, 1)
    s7End = datetime.time(16, 15, 0)


    if time_in_range(s1Start, s1End, ts):
      timecat = 1 
    elif time_in_range(s2Start, s2End, ts):
      timecat = 2 
    elif time_in_range(s3Start, s3End, ts):
      timecat = 3 
    elif time_in_range(s4Start, s4End, ts):
      timecat = 4 
    elif time_in_range(s5Start, s5End, ts):
      timecat = 5 
    elif time_in_range(s6Start, s6End, ts):
      timecat = 6 
    elif time_in_range(s7Start, s7End, ts):
      timecat = 7 

    return timecat

def time_in_range(start, end, x):
    if start <= end:
        return start <= x <= end
    else:
        return start <= x or x <= end
```

```python
for index,row in tatasteel_5min.iterrows():
    tatasteel_5min.loc[index,'timerange'] = getTimeCat(row['Date/Time'])

tatasteel_5min['Date/Time'] = pd.to_datetime(tatasteel_5min['Date/Time'],format="%d-%m-%Y %H:%M:%S")
tatasteel_5min['day_of_week'] = tatasteel_5min['Date/Time'].dt.dayofweek
tatasteel_5min['timerange'] = tatasteel_5min['timerange'].astype(int)
tatasteel_5min.head()
```

# Creating a Sliding Window and splitting into X and y variables

```python
#Creating a data structure that takes in 24 timesteps for the formation of each datapoint in X
#And also takes in the 48th value of the target_class column for the formation of each datapoint in y 

X = []
y = []

for i in range (24,len(tatasteel_merged)-24):
    X.append(np.array(tatasteel_merged.iloc[i-24:i,0:171]))
    y.append(np.array(tatasteel_merged.iloc[i+24,171]))

X, y = np.array(X), np.array(y)

print(X.shape)
print(y.shape)
```

**Re-Shaping windowed data from 3d array to 2d array**

```python
#To reshape to 2d array: (n_datapoints.n_rows,(n_timesteps x n_features))
X = np.reshape(X, (X.shape[0], X.shape[1]*X.shape[2]))
print(X.shape)
```

# Scaling the Data

```python
X = pd.DataFrame(X)

from scipy.stats import zscore
X_z = X.apply(zscore)
X_z
```

# Splitting into Train and Test Sets with a ratio of 80:20

```python
X_train = X_z[0:11250]
y_train = y[0:11250]
X_test = X_z[11250:]
y_test = y[11250:]

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Random Forest Model


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import classification_report

rf_model = RandomForestClassifier()

rf_model.fit(X_train,y_train)

rf_pred_train = rf_model.predict(X_train)
rf_pred_test = rf_model.predict(X_test)


print("Train set accuracy =", rf_model.score(X_train,y_train))
print("Test set accuracy  =", rf_model.score(X_test,y_test))
print("_______________________________________________________")
print("Confusion matrix:")
print(metrics.confusion_matrix(y_test,rf_pred_test))
print("_______________________________________________________")
print("Classification report:")
print(classification_report(y_test,rf_pred_test))
```
